```python
import matplotlib.pyplot as plt
import matplotlib.style
import pandas as pd
import numpy as np


matplotlib.style.use('ggplot')
```

```python
df = pd.read_csv('http://bit.ly/kaggletrain')
```

```python
df
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15,15))
x=[1,2]
y=df[df.loc[:, "Sex"]=="male"].count()
z=df[df.loc[:, "Sex"]=="female"].count()
a=[y.Sex,z.Sex]
labels = ['male', 'female']
axes[0,0].bar(x, a, tick_label=labels)



y=df["Age"]
axes[0,1].hist(y, bins=20)


z = [1, 2, 3]
c1=df[(df.loc[:, "Pclass"]== 1)& (df.loc[:, "Survived"]== 0)].count()
c2=df[(df.loc[:, "Pclass"]== 2)& (df.loc[:, "Survived"]== 0)].count()
c3=df[(df.loc[:, "Pclass"]== 3)& (df.loc[:, "Survived"]== 0)].count()
c=[c1.Survived,c2.Survived,c3.Survived]
d1=df[(df.loc[:, "Pclass"]== 1)& (df.loc[:, "Survived"]== 1)].count()
d2=df[(df.loc[:, "Pclass"]== 2)& (df.loc[:, "Survived"]== 1)].count()
d3=df[(df.loc[:, "Pclass"]== 3)& (df.loc[:, "Survived"]== 1)].count()
d=[d1.Survived,d2.Survived,d3.Survived]
labels2 = ['Pclass:1', 'Pclass:2', 'Pclass:3']
width = 0.4
z2 = [num + width for num in z]
axes[1,0].bar(z , c, width=width, tick_label=labels2,label='Dead')
axes[1,0].bar(z2 , d, width=width, label='Live')
axes[1,0].legend()




die=df[df.loc[:, "Survived"]== 0]
live=df[df.loc[:, "Survived"]== 1]
axes[1,1].scatter(die.Age, die.Fare, marker='x',
           label='x')
axes[1,1].scatter(live.Age, live.Fare, marker='s',
           label='square')

plt.show()
```
